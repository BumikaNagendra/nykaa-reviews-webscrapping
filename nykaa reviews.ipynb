```python
html = driver.page_source.encode('utf-8')
df=pd.DataFrame()
#for i in range(1,10):
#url="https://www.nykaa.com/nykaa-skin-secrets-sheet-mask/reviews/418274?ptype=reviews&skuId=494580"
#data = requests.get(url)
soup = BeautifulSoup(data.content, 'html.parser')
#category=url.split('/')[3]
parent_tags=soup.findAll('div',{'class':'review-body'})


for parent in parent_tags:
        try:
            date=parent.find('div',{'class':'review-date'}).text
        except:
            date=None
        try:
            review_headline=parent.find('h4',{'class':'title'}).text
        except:
            review_headline=None
    
        try:
            reviews=parent.find('p',{'class':'description'}).text
        except:
            reviews=None
    
        try:
            ratings=parent.find('div',{'class':'rating green-bck'}).text
        except:
            ratings=None
    
        temp={'date':date,
        'review_headline':review_headline,
        'reviews':reviews,
        'ratings':ratings,
        'category':category}
        
        df=df.append(temp, ignore_index=True)
    
df[['date','review_headline','reviews','ratings','category']]
```
