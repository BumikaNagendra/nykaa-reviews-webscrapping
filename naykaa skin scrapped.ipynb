```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

%matplotlib inline
```

```python
url="https://www.nykaa.com/skin/c/8377?root=nav_1&dir=desc&order=popularity"
category=url.split('/')[3]
data = requests.get(url)
#data.content
soup = BeautifulSoup(data.content)
parent_tags=soup.findAll('div',{'class':'card-wrapper-container col-xs-12 col-sm-6 col-md-4'})
len(parent_tags)
parent_tags
```

```python
df=pd.DataFrame()
for parent in parent_tags:
    try:
        product_name=parent.find('div',{'class':'m-content__product-list__title'}).text
    except:
        product_name=None
    try:
        mrp=parent.find('span',{'class':'post-card__content-price-offer'}).text
    except:
        mrp=None
    try:
        discount=parent.find('span',{'class':'post-card__offers-offer'}).text
    except:
        discount=None
    try:
        ratings=parent.find('div',{'class':'m-content__product-list__ratings'}).text
    except:
        ratings=None
    
    temp={
        'product_name':product_name,
        'mrp':mrp,
        'discount':discount,
        'ratings':ratings,
        'category':category
    }
    df=df.append(temp, ignore_index=True)
df[['product_name','mrp','discount','ratings','category']].dropna(how='all')
```
